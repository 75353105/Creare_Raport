```python
#Eliminare coloane 'Unnamed'
website_dataset = website_dataset.loc[:, ~website_dataset.columns.str.contains('Unnamed')]
google_dataset = google_dataset.loc[:, ~google_dataset.columns.str.contains('Unnamed')]
facebook_dataset= facebook_dataset.loc[:, ~facebook_dataset.columns.str.contains('Unnamed')]
```

```python
print(google_dataset.columns)
print(website_dataset.columns)
print(facebook_dataset.columns)
```

```python
#Schimbarea denumirilor coloanelor principale
facebook_dataset = facebook_dataset.rename(columns={'categories' : 'category'})
website_dataset = website_dataset.rename(columns={'s_category' : 'category'})
website_dataset = website_dataset.rename(columns={'root_domain' : 'domain'})
website_dataset = website_dataset.rename(columns={'main_region' : 'region_name'})
website_dataset = website_dataset.rename(columns={'main_country' : 'country_name'})
website_dataset = website_dataset.rename(columns={'site_name' : 'name'})
website_dataset = website_dataset.rename(columns={'main_city' : 'city'})
website_dataset = website_dataset.rename(columns={'tld' : 'country_code'})
```

```python
print(google_dataset.columns)
print(website_dataset.columns)
print(facebook_dataset.columns)
```

```python
#Selectam coloanele comune din fiecare set de date
common_columns = ['domain', 'name' ,'phone', 'city', 'country_name', 'region_name', 'category', 'country_code']

#Cream o jonciune externa cu ajutorul coloane domain pentru a putea pastra toate datele importante
complete_dataset = pd.merge(facebook_dataset, google_dataset, on='domain', how='outer', suffixes=('_fb', '_gg'))
complete_dataset = pd.merge(complete_dataset, website_dataset, on='domain', how='outer', suffixes=('', '_ws'))

#Pentru o lizibilitate mai buna sortam coloanele in noul set de date
final_columns = common_columns + [col for col in complete_dataset.columns if col not in common_columns]
complete_dataset = complete_dataset[final_columns]

#Completam valorile lipsa in coloanele comune
complete_dataset['name'] = complete_dataset['name_fb'].fillna(complete_dataset['name_gg']).fillna(complete_dataset['name'])
complete_dataset['category'] = complete_dataset['category_fb'].fillna(complete_dataset['category_gg']).fillna(complete_dataset['category'])
complete_dataset['phone'] = complete_dataset['phone_fb'].fillna(complete_dataset['phone_gg']).fillna(complete_dataset['phone'])
complete_dataset['city'] = complete_dataset['city_fb'].fillna(complete_dataset['city_gg']).fillna(complete_dataset['city'])
complete_dataset['country_name'] = complete_dataset['country_name_fb'].fillna(complete_dataset['country_name_gg']).fillna(complete_dataset['country_name'])
complete_dataset['region_name'] = complete_dataset['region_name_fb'].fillna(complete_dataset['region_name_gg']).fillna(complete_dataset['region_name'])
complete_dataset['country_code'] = complete_dataset['country_code_fb'].fillna(complete_dataset['country_code_gg']).fillna(complete_dataset['country_code'])

print(complete_dataset.head(10))
```

```python
#Setul a fost creat cu succes, insa acuma trebuie optimizat si actualizat astfel incat el sa poata fi cat mai functional
print(complete_dataset.columns)
```

```python
#Se vor analiza care coloane sunt si care sunt redundante. Coloanele comune cu sufix precum category_gg sau phone_fb
#se pot considera deja redundante si pot fi eliminate
complete_dataset = complete_dataset.drop(columns=['name_fb','name_gg','category_fb','category_gg','phone_gg','phone_fb','city_fb','city_gg',
                               'country_name_fb','country_name_gg','region_name_fb','region_name_gg','country_code_fb','country_code_gg',
                                                 'region_code_fb','region_code_gg'])
```

```python
complete_dataset.columns
```

```python
print(complete_dataset.head(10))
```

```python
#Observam ca anumite coloane ramase sunt goale, dar ar putea fi completate cu restul de informatii necesare.
complete_dataset['description'] = complete_dataset['description'].fillna(facebook_dataset['description'])
complete_dataset['email'] = complete_dataset['email'].fillna(facebook_dataset['email'])
complete_dataset['link'] = complete_dataset['link'].fillna(facebook_dataset['link'])
complete_dataset['address_fb'] = complete_dataset['address_fb'].fillna(facebook_dataset['address'])
complete_dataset['zip_code_fb'] = complete_dataset['zip_code_fb'].fillna(facebook_dataset['zip_code'])
complete_dataset['phone_country_code_fb'] = complete_dataset['phone_country_code_fb'].fillna(facebook_dataset['phone_country_code'])
complete_dataset['page_type'] = complete_dataset['page_type'].fillna(facebook_dataset['page_type'])
complete_dataset['address_gg'] = complete_dataset['address_gg'].fillna(google_dataset['address'])
complete_dataset['zip_code_gg'] = complete_dataset['zip_code_gg'].fillna(google_dataset['zip_code'])
complete_dataset['phone_country_code_gg'] = complete_dataset['phone_country_code_gg'].fillna(google_dataset['phone_country_code'])
complete_dataset['legal_name'] = complete_dataset['legal_name'].fillna(website_dataset['legal_name'])
complete_dataset['language'] = complete_dataset['language'].fillna(website_dataset['language'])
complete_dataset['domain_suffix'] = complete_dataset['domain_suffix'].fillna(website_dataset['domain_suffix'])
complete_dataset.head(15)
```

```python
#Eliminam coloanele care nu ofera suficienta relevanta setului de date
complete_dataset = complete_dataset.drop(columns = ['phone_country_code_fb','phone_country_code_gg','zip_code_fb',
                                                    'zip_code_gg'])
complete_dataset.head(15)
```

```python
#Cum acuma avem setul de date cu variabilele relevante, va trebui sa cautam conflictele de date
#Duplicatele
duplicates = complete_dataset[complete_dataset.duplicated(subset='domain', keep=False)]
print(duplicates)
```

```python
#Avem duplicate, asa ca vom cauta randurile cele mai putin complete si le vom elimina
complete_dataset = complete_dataset.sort_values(by='name').drop_duplicates(subset='domain', keep='last')
```

```python
#Verificam valorile lipsa
missing_data = complete_dataset[complete_dataset[['name', 'phone','category','country_name',
                                                  'country_code','region_name']].isnull().any(axis=1)]
print(missing_data)
```

```python
#Inlocuim valorile categorice cu un text backup
complete_dataset['name'].fillna(method='ffill', inplace=True)
complete_dataset['phone'].fillna(method='bfill', inplace=True)
complete_dataset['category'].fillna(method='ffill', inplace=True)
complete_dataset['country_name'].fillna(method='bfill', inplace=True)
complete_dataset['country_code'].fillna(method='ffill', inplace=True)
complete_dataset['country_code'].fillna(method='bfill', inplace=True)
```
